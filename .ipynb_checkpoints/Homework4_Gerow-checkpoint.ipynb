Exercise 1: Write a function that computes the level of output generated by a constant returns to scale Cobb-Douglas production function, i.e., such that it computes  ùê¥‚ãÖùêæùõº‚ãÖùêø1‚àíùõº .

```python
def my_CobbDouglas(K=1, L=1, A=1, alpha=0.3):
    '''
    This function computes the level of output of a Cobb-Douglas production function.
    Given values of K, L, A, and alpha it returns $A\cdot K^\alpha \cdot L^{1-\alpha}$.
    Parameters:
    -----------
    K: float, integer or any other number, default=1
    L: float, integer or any other number, default=1
    A: float, integer or any other number, default=1
    alpha: float between (0,1), default=0.3
    
    Returns:
    --------
    out = A * K**alpha * L**(1-alpha)
    
    Example:
    >>> out = CobbDouglas(4, 9, 2, 1/2)
    >>> out
        12
    '''
    return A * K**alpha * L**(1-alpha) 
```

```python
my_CobbDouglas(1,1,1,.3)
```

```python
my_CobbDouglas(4,9,2,.5)
```

```python
K = np.linspace(0, 5, 100)
print(K)
```

Exercise 2: Use the function you created and plot the production function as a function of  ùêæ  for given values of  ùê¥  and  ùêø . Hint: Use the np.linspace(0, 5, 100) function to create an array of values of  ùêæ  with 100 point between 0 and 5.

```python
x = my_CobbDouglas(K)
plt.plot(x, c='r', marker='o', markersize=4, linestyle='-')
plt.xlabel(r'input', fontsize=18)
plt.ylabel(r'output', fontsize=18)
plt.title(r'Cobb Douglas Graph', fontsize=24)
plt.legend(fontsize=18)
plt.savefig('Exercise2.png')
plt.savefig('Exercise2.pdf')
plt.savefig('Exercise2.jpeg')
```
